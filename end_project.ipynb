### MÃ©thodes d'identification des valeurs aberrantes

```python
# Variables numÃ©riques Ã  analyser
numeric_columns = ["cpu_frequency", "gpu_frequency", "price", "ram", "ram_frequency", "hdd", "ssd", "screen_size"]

# CrÃ©ation de boxplots pour visualiser les valeurs aberrantes
plt.figure(figsize=(16, 12))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(df1[col])
    plt.title(f"Boxplot de {col}")

plt.tight_layout()
plt.show()

# DÃ©tection des valeurs aberrantes avec l'IQR
outliers = {}
for col in numeric_columns:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[col] = df1[(df1[col] < lower_bound) | (df1[col] > upper_bound)][col]

# RÃ©sumÃ© des valeurs aberrantes
outliers_summary = {col: len(outliers[col]) for col in numeric_columns}
print("Nombre de valeurs aberrantes dÃ©tectÃ©es par variable :")
print(outliers_summary)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Initialisation des modÃ¨les
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lasso Regression": Lasso(alpha=0.1),
    "Decision Tree": DecisionTreeRegressor(max_depth=10, random_state=42)
}

# Dictionnaire pour stocker les performances des modÃ¨les
model_performance = {}

# Fonction pour Ã©valuer le modÃ¨le
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)

    # Calcul des mÃ©triques
    mae = mean_absolute_error(y_test, predictions)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    r2 = r2_score(y_test, predictions)

    return mae, rmse, r2

# Ã‰valuation de chaque modÃ¨le
for model_name, model in models.items():
    mae, rmse, r2 = evaluate_model(model, X_train, y_train, X_test, y_test)
    model_performance[model_name] = {
        "MAE": mae,
        "RMSE": rmse,
        "R2 Score": r2
    }

# Affichage des performances
for model_name, metrics in model_performance.items():
    print(f"\nModÃ¨le : {model_name}")
    print(f"MAE : {metrics['MAE']:.2f}")
    print(f"RMSE : {metrics['RMSE']:.2f}")
    print(f"RÂ² Score : {metrics['R2 Score']:.2f}")
```
